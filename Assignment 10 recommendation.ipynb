```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
df=pd.read_csv('Book.csv',encoding='Latin1') 
df 
```

```python
# converting long data into wide data using pivot table
df2=df1.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
df2 
```

```python
# Replacing the index values by unique user Ids
df2.index=df1['User.ID'].unique()
df2 
```

```python
# convering null values with 0 
df2.fillna(0,inplace=True)
df2 
```

```python
# using Cosine Similarity 
cos=1-pairwise_distances(df2.values,metric='cosine')
cos 
```

```python
# convering above result into data frame 
cos1=pd.DataFrame(cos)
cos1 
```

```python
# assigning  index and column names to user ids 
cos1.index=df1['User.ID'].unique()
cos1.columns=df1['User.ID'].unique()
cos1 
```

```python
# Nullifying diagonal values so we don't confuse
np.fill_diagonal(cos,0)
cos1 
```

```python
#  Users having same rating 
cos1.idxmax(axis=1) 
```
